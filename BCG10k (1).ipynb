```python
# Step 1: Import pandas
import pandas as pd
```

```python
# Step 2: Load your data
msft = pd.read_csv('Microsoft_Financials.csv')
tesla = pd.read_csv('Tesla_Financials.csv')
apple = pd.read_csv('Apple_Financials.csv')

# Step 3: Add a company name column
msft['Company'] = 'Microsoft'
tesla['Company'] = 'Tesla'
apple['Company'] = 'Apple'
```

```python
# Step 4: Combine all three datasets into one
df = pd.concat([msft, tesla, apple], ignore_index=True)

# Step 5: Sort by Company and Year to prepare for trend analysis
df = df.sort_values(by=['Company', 'Year'])
```

```python
# Step 6: Analyze trends with pandas
df['Revenue Growth %'] = df.groupby('Company')['Total Revenue'].pct_change() * 100
df['Net Income Growth %'] = df.groupby('Company')['Net Income'].pct_change() * 100

# Step 7: View the result
df[['Company', 'Year', 'Total Revenue', 'Revenue Growth %', 'Net Income', 'Net Income Growth %']]
```

```python
df.to_csv('Combined_Financial_Analysis.csv', index=False)
```
